TASK - B

```python
# Define the matrix A
A = np.array([[1 + 0.05, -0.05],
              [0.05, 1 - 0.08]])

# Initial populations
p_current = np.array([8000, 2000])  # 8000 rabbits, 2000 foxes

# Lists to store population data for plotting
rabbit_populations = [p_current[0]]
fox_populations = [p_current[1]]

# Simulate population dynamics for 100 time steps
for _ in range(100):
    p_current = np.dot(A, p_current)
    rabbit_populations.append(p_current[0])
    fox_populations.append(p_current[1])

# Calculate relative populations as percentages
total_populations = np.array(rabbit_populations) + np.array(fox_populations)
relative_rabbit_populations = np.array(rabbit_populations) / total_populations * 100
relative_fox_populations = np.array(fox_populations) / total_populations * 100

# Generate line graph
time_steps = np.arange(101)  # 0 to 100 time steps
plt.plot(time_steps, relative_rabbit_populations, label='Rabbits')
plt.plot(time_steps, relative_fox_populations, label='Foxes')

# Set labels and legend
plt.xlabel('Time Steps')
plt.ylabel('Relative Population (%)')
plt.title('Population Dynamics of Rabbits and Foxes')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the modified matrix A
A_modified = np.array([[1 + 0.05, -0.2],
                       [0.05, 1 - 0.08]])

# Function to simulate population dynamics for a given initial population
def simulate_population_dynamics(initial_population):
    p_current = initial_population
    rabbit_trajectory = [p_current[0]]
    fox_trajectory = [p_current[1]]

    # Simulate population dynamics for 100 time steps
    for _ in range(100):
        p_current = np.dot(A_modified, p_current)
        rabbit_trajectory.append(p_current[0])
        fox_trajectory.append(p_current[1])

    return rabbit_trajectory, fox_trajectory

# Generate trajectories for different initial populations
initial_populations = [
    np.array([8000, 2000]),  # Initial populations from part (b)
    np.array([5000, 1000]),
    np.array([10000, 2500]),
]

# Plot the trajectories
for initial_population in initial_populations:
    rabbit_trajectory, fox_trajectory = simulate_population_dynamics(initial_population)
    plt.plot(rabbit_trajectory, fox_trajectory, label=f'Initial: {initial_population}')

# Set labels and legend
plt.xlabel('Rabbit Population')
plt.ylabel('Fox Population')
plt.title('Population Dynamics of Rabbits and Foxes with Modified Matrix')
plt.legend()

# Show the plot
plt.show()
```
